```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("imdb_top_1000.csv")
```

```python
df_nl = df.isnull().sum()
```

```python
df_nl

#missing values are certificate,Meta_score,No_of_votesb
```

```python
df_nl.plot(kind='bar')
plt.xlabel("columns")
plt.ylabel("missing values")
plt.title("missing data per column")
plt.show()
```

```python
df['Certificate'].isnull().sum()
df['Certificate'].mode()
```

```python
mode_value = df['Certificate'].mode()[0]

```

```python
df['Certificate'].fillna(mode_value,inplace=True)
```

```python
df['Certificate'].isnull().sum()


#replaced missing values with mode of that column
```

```python
df['Meta_score'].isnull().sum()
mean_val = df['Meta_score'].mean()
df['Meta_score'].fillna(mean_val,inplace=True)
```

```python
df['Meta_score'].isnull().sum()
```

```python
df['Gross'] = pd.to_numeric(df['Gross'],errors='coerce')
```

```python
df['Gross'].isnull().sum()
mn_val = df["Gross"].mean()
df['Meta_score'].fillna(mn_val,inplace=True)
```
